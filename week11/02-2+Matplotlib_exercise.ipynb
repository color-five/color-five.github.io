## Matplotlib练习

### 1. 用蓝线绘制函数𝑓(𝑥) = 𝑥^2−2𝑥+3的图表，并在坐标(1,2)位置处标上一个红色的点


```python
import numpy as np
import matplotlib.pyplot as plt
#定义x的范围
x=np.linspace(-8,8,400)
#写出函数表达式
f = x*x-2*x+3
#画出函数图
plt.plot(x,f,color="blue")
plt.plot(1,2,"ro")
```

### 2. 使用np.linspace()函数生成t∈[0, 2𝜋]。给定𝑥=16*sin⁡(𝑡)^3和𝑦=13*cos⁡(𝑡)−5*cos⁡(2𝑡)−2*cos⁡(3𝑡)−cos⁡(4𝑡)，画出x-y图表，并给图标添加一个题目：Heart

```python
import numpy as np
import matplotlib.pyplot as plt
#定义t的范围
t=np.linspace(0,2*np.pi,8000)
#写出函数表达式
x = 16*np.sin(t)*np.sin(t)*np.sin(t)
y = 13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-4*np.cos(4*t)
#画出函数图
plt.plot(x,y)
plt.title('Heart')
```
