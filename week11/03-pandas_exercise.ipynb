## Step 1: 导入可能需要的库

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

## Step 2: 读取数据集

```python
titanic = pd.read_csv('titanic.csv')
```

```python
# 请填写相应代码
print(titanic.shape)
```

## Step 5: 将PassengerID设置为索引

```python
titanic.set_index('PassengerId')
```

## Step 6：数据中有缺失值吗？

```python
# 请填写相应代码
# 统计每一列的缺失值数量
print(titanic.isnull().sum())
```

## Step 7: 乘客的最大年龄和最小年龄是多少？

```python
# 请填写相应代码
max_age = titanic['Age'].max()
```

```python
# 请填写相应代码
min_age = titanic['Age'].min()
print(f"最大年龄: {max_age}, 最小年龄: {min_age}")
```

## Step 8: 有多少人生还？

```python
# 请填写相应代码
survived_count = titanic['Survived'].sum()
print(f"生还人数: {survived_count}")
```

## Step 9: 男性和女性的生还比例分别是多少？

```python
# 请填写相应代码
# 使用 pivot_table 计算不同性别（Sex）下的生还（Survived）平均值
survival_rate_by_sex = titanic.pivot_table(values='Survived', index='Sex', aggfunc='mean')
```

```python
# 请填写相应代码
print(survival_rate_by_sex)
```

## Step 10: 按照船票价格降序排列

```python
# 请填写相应代码
# ascending=False 表示降序
sorted_titanic = titanic.sort_values(by='Fare', ascending=False)
print(sorted_titanic[['Name', 'Fare']].head()) # 打印前几行查看结果
```

## Step 11: 绘制一个展示船票价格的直方图

```python
# 请填写相应代码
plt.figure(figsize=(10, 6))
plt.hist(titanic['Fare'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

## Step 12: 绘制乘客年龄与船票价格的散点图

```python
# 请填写相应代码
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5, color='green')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
