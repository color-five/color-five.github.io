```python
# 假设这是给定的整数列表 L
L = [1, 2, 3, 4, 5] 

# 获取 L 的升序排列
sorted_L = sorted(L)

if L == sorted_L:
    # 如果原列表等于升序排列后的列表，说明是升序
    print("UP")
elif L == sorted_L[::-1]:
    # 如果原列表等于升序排列的倒序（即降序），说明是降序
    # [::-1] 是 Python 中常用的切片倒序写法
    print("DOWN")
else:
    # 既不是升序也不是降序，说明是无序
    print("WRONG")
```

### 1.6. 互联网上的每台计算机都有一个IP，合法的IP格式为：A.B.C.D。其中A、B、C、D均为[0, 255]中的整数。为了简单起见，我们规定这四个整数中不允许有前导零存在，如001。现在给你一个字符串s（s不含空白符），请你判断s是不是合法IP，若是，输出Yes，否则输出No。如：s=“202.115.32.24”，则输出Yes；s=“a.11.11.11”, 则输出No。

```python
def is_valid_ip(s):
    # 1. 用点号分割字符串，必须刚好分成4段
    parts = s.split('.')
    if len(parts) != 4:
        return False
    
    # 2. 遍历每一段进行校验
    for part in parts:
        # 检查是否为空字符串，或者是否包含非数字字符
        if not part or not part.isdigit():
            return False
        
        # 检查前导零：如果长度大于1且以'0'开头，则非法（如 "01", "001"）
        if len(part) > 1 and part[0] == '0':
            return False
        
        # 转换为整数，检查是否在 [0, 255] 范围内
        num = int(part)
        if num < 0 or num > 255:
            return False
            
    return True

# 测试样例
s1 = "202.115.32.24"
s2 = "a.11.11.11"
s3 = "192.168.01.1"  # 包含前导零，应为 No
s4 = "256.1.1.1"     # 超出255，应为 No

for s in [s1, s2, s3, s4]:
    if is_valid_ip(s):
        print(f"s=\"{s}\"，则输出Yes")
    else:
        print(f"s=\"{s}\"，则输出No")
```

### 2.2.2 继承：创建一个 Student 类，继承自 Person 类，并添加属性 student_id 和方法 study()。

```python
# 1. 首先定义父类 Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"大家好，我叫 {self.name}，今年 {self.age} 岁。")

# 2. 定义子类 Student，继承自 Person
class Student(Person):  # 括号里写上父类名，表示继承
    def __init__(self, name, age, student_id):
        # 调用父类的 __init__ 方法，初始化继承来的 name 和 age
        super().__init__(name, age)
        # 添加子类特有的属性 student_id
        self.student_id = student_id

    # 添加子类特有的方法
    def study(self):
        print(f"{self.name}（学号：{self.student_id}）正在认真学习。")

# 3. 实例化 Student 对象并测试
student1 = Student("小明", 18, "2024001")

# 调用从父类继承来的方法
student1.introduce()  # 输出: 大家好，我叫小明，今年 18 岁。
# 调用子类特有的方法
student1.study()      # 输出: 小明（学号：2024001）正在认真学习。
```

### 2.2.4 类方法与静态方法：创建一个 MathUtils 类，包含一个类方法 add() 和一个静态方法 multiply()

```python
class MathUtils:
    # 定义类方法，需要使用 @classmethod 装饰器，且第一个参数必须是 cls
    @classmethod
    def add(cls, a, b):
        return a + b

    # 定义静态方法，需要使用 @staticmethod 装饰器，不需要特殊的第一个参数
    @staticmethod
    def multiply(a, b):
        return a * b

# 调用类方法（推荐通过类名直接调用）
result_add = MathUtils.add(5, 3)
print(f"加法结果: {result_add}")  # 输出: 加法结果: 8

# 调用静态方法（推荐通过类名直接调用）
result_multiply = MathUtils.multiply(5, 3)
print(f"乘法结果: {result_multiply}")  # 输出: 乘法结果: 15
```

### 2.2.5 属性装饰器：创建一个 Circle 类，包含属性 radius，并通过属性装饰器确保半径不能为负数。

```python
class Circle:
    def __init__(self, radius):
        # 注意：这里调用 self.radius 会触发下面的 @radius.setter 方法
        self.radius = radius  

    # 1. 定义获取 radius 的方法（Getter）
    @property
    def radius(self):
        return self._radius  # 返回实际存储半径的内部变量

    # 2. 定义设置 radius 的方法（Setter），并加入验证逻辑
    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("半径不能为负数！")  # 如果值为负，抛出异常
        self._radius = value  # 只有通过验证的值，才会被存入内部变量 _radius

# 正常实例化对象
circle = Circle(5)
print(circle.radius)  # 输出: 5

# 尝试修改为合法的正数
circle.radius = 10
print(circle.radius)  # 输出: 10

# 尝试设置为负数，会触发 ValueError 异常
try:
    circle.radius = -3
except ValueError as e:
    print(e)  # 输出: 半径不能为负数！
```

### 2.2.7 魔术方法（特殊函数）：创建一个 Point 类，有x和y两个属性，表示二维坐标系中的点，并实现 __add__ 和 __str__ 魔术方法。

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # 1. 实现 __add__ 魔术方法，定义两个点相加的行为
    def __add__(self, other):
        # self 代表加号左边的点，other 代表加号右边的点
        return Point(self.x + other.x, self.y + other.y)

    # 2. 实现 __str__ 魔术方法，定义打印对象时的字符串显示形式
    def __str__(self):
        return f"Point({self.x}, {self.y})"

# 实例化两个点
p1 = Point(1, 2)
p2 = Point(3, 4)

# 触发 __add__ 魔术方法
p3 = p1 + p2
print(p3)  # 触发 __str__ 魔术方法，输出: Point(4, 6)
```

### 2.2.8 单例模式：实现一个单例模式的 Logger 类，确保只有一个实例。

```python
class Logger:
    # 定义一个私有类属性，用来保存唯一的实例，初始为 None
    _instance = None

    def __new__(cls, *args, **kwargs):
        # 判断是否已经创建过实例
        if cls._instance is None:
            # 如果没有，就调用父类的 __new__ 创建一个新的实例并保存
            cls._instance = super().__new__(cls)
        # 如果已经创建过，就直接返回之前保存的那个实例
        return cls._instance

    def __init__(self, name="AppLogger"):
        # 初始化日志记录器的名称等配置
        # 注意：在严格的单例中，__init__ 可能会被多次调用，
        # 实际开发中通常会在这里加一个标志位防止重复初始化。
        self.name = name

    def log(self, message):
        print(f"[{self.name}] Log: {message}")

# --- 测试代码 ---
# 第一次创建实例
logger1 = Logger("SystemLogger")
logger1.log("程序启动")  # 输出: [SystemLogger] Log: 程序启动

# 第二次尝试创建实例
logger2 = Logger("UserLogger")
logger2.log("用户登录")  # 输出: [SystemLogger] Log: 用户登录

# 验证两个变量是否指向同一个内存地址（即同一个对象）
print(logger1 is logger2)  # 输出: True
```

### 2.2.9 抽象基类：创建一个抽象基类 Shape，并实现两个子类 Circle 和 Rectangle，要求每个子类都必须实现 area() 方法。

```python
from abc import ABC, abstractmethod

# 1. 定义抽象基类 Shape，继承自 ABC
class Shape(ABC):
    # 2. 使用 @abstractmethod 装饰器定义抽象方法 area
    @abstractmethod
    def area(self):
        pass  # 抽象方法不需要具体的实现代码

# 3. 实现子类 Circle，继承自 Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    # 必须实现 area 方法，否则会报错
    def area(self):
        return 3.14159 * self.radius * self.radius

# 4. 实现子类 Rectangle，继承自 Shape
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # 必须实现 area 方法，否则会报错
    def area(self):
        return self.width * self.height

# --- 测试代码 ---
# 正常实例化具体的子类
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"圆形的面积: {circle.area()}")      # 输出: 圆形的面积: 78.53975
print(f"矩形的面积: {rectangle.area()}")  # 输出: 矩形的面积: 24

# 尝试直接实例化抽象基类 Shape，会直接抛出 TypeError 报错
# shape = Shape()  # 报错: TypeError: Can't instantiate abstract class Shape with abstract method area
```

### 2.3.1 以列表存储矩阵，编写一个函数，计算两个矩阵的乘积

```python
def matrix_multiply(matrix_a, matrix_b):
    # 1. 获取两个矩阵的维度
    rows_a = len(matrix_a)          # 矩阵A的行数
    cols_a = len(matrix_a[0])       # 矩阵A的列数
    rows_b = len(matrix_b)          # 矩阵B的行数
    cols_b = len(matrix_b[0])       # 矩阵B的列数

    # 2. 检查维度是否满足相乘条件
    if cols_a != rows_b:
        raise ValueError(f"矩阵维度不匹配，无法相乘：矩阵A的列数({cols_a}) 不等于 矩阵B的行数({rows_b})")

    # 3. 初始化结果矩阵（全为 0 的 m×p 矩阵）
    # 注意：必须使用列表推导式为每一行创建独立的列表，避免浅拷贝导致的行数据同步修改问题
    result = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    # 4. 三重循环计算矩阵乘积
    for i in range(rows_a):             # 遍历矩阵A的每一行
        for j in range(cols_b):         # 遍历矩阵B的每一列
            for k in range(cols_a):     # 遍历A的列 / B的行，进行对应元素相乘并累加
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]
    
    return result

# --- 测试代码 ---
# 定义两个测试矩阵 (2x3) * (3x2) = (2x2)
A = [
    [1, 2, 3],
    [4, 5, 6]
]
B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

try:
    C = matrix_multiply(A, B)
    print("矩阵相乘的结果为：")
    for row in C:
        print(row)
except ValueError as e:
    print(e)
```
