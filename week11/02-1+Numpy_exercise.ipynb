### 1. 创建一个长度为10全为0的一维数组，然后让第5个元素等于1。

```python
import numpy as np

# 创建一个长度为10、全为0的一维数组
x = np.zeros(10)

# 让第5个元素（索引为4）等于1
x[4] = 1

# 打印结果
print(x)
```

### 2. 生成一个4X5和一个5X4的矩阵，并计算它们的乘积。

```python
import numpy as np
#创建矩阵A和B
A = np.array([[1,2,3,4],[0,0,0,1],[3,7,8,9],[8,9,0,1],[1,1,1,1]])
B = np.array([[1,2,5,6,0],[5,5,5,5,9],[3,5,1,1,1],[2,2,2,2,2]])
#计算乘积
C = A@B
#输出结果
print(C)
```

### 4. 随机生成一个5行10列的矩阵，然后每行元素减去该行的平均值

```python
import numpy as np
#随机生成一个5x10的矩阵
A = np.random.randint(0,100,size=(5,10))
#计算每一行的平均值
B = A.mean(axis=1,keepdims=True)
#计算结果
C = A - B
#输出结果
print(C)
```
