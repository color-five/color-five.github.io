```python
import torch

def matrix_multiply_torch(A, B):
    """
    使用PyTorch实现矩阵乘法。
    Args:
        A: 类数组结构，可转换为torch.Tensor，形状 (m, n)
        B: 类数组结构，可转换为torch.Tensor，形状 (n, p)
    Returns:
        C: torch.Tensor，乘积矩阵，形状 (m, p)
    """
    A_t = torch.tensor(A, dtype=torch.float32)
    B_t = torch.tensor(B, dtype=torch.float32)
    
    # 方法1: 使用 torch.mm (专门用于2D矩阵)
    C_mm = torch.mm(A_t, B_t)
    # 方法2: 使用 torch.matmul (通用，支持高维)
    C_matmul = torch.matmul(A_t, B_t)
    # 方法3: 使用 @ 运算符
    C_at = A_t @ B_t
    
    # 验证结果一致
    assert torch.allclose(C_mm, C_matmul) and torch.allclose(C_matmul, C_at), "PyTorch计算结果不一致"
    return C_mm

# 测试
A_t = [[1, 2], [3, 4]]
B_t = [[5, 6], [7, 8]]
C_t = matrix_multiply_torch(A_t, B_t)
print("PyTorch计算结果：\n", C_t)  # 输出：tensor([[19., 22.], [43., 50.]])
```
