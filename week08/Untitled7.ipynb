```python
def binary_confusion_matrix(y_true, y_pred):
    """
    计算二分类混淆矩阵。
    参数:
        y_true: list 或 array-like, 真实标签 (0 或 1)。
        y_pred: list 或 array-like, 预测标签 (0 或 1)。
    返回:
        cm: list of lists, 2x2混淆矩阵，格式为 [[TN, FP], [FN, TP]]。
    """
    # 初始化2x2矩阵
    cm = [[0, 0], [0, 0]]
    
    # 确保输入长度一致
    if len(y_true) != len(y_pred):
        raise ValueError("y_true 和 y_pred 的长度必须相同")
    
    for true, pred in zip(y_true, y_pred):
        # 将标签映射到矩阵索引（0或1）
        cm[true][pred] += 1
    
    return cm

# 示例使用
y_true = [0, 1, 0, 1, 0, 1, 0, 1]
y_pred = [0, 0, 1, 1, 0, 1, 0, 1]
cm_binary = binary_confusion_matrix(y_true, y_pred)
print("二分类混淆矩阵:")
print(f"TN: {cm_binary[0]}, FP: {cm_binary[1]}")
print(f"FN: {cm_binary[0]}, TP: {cm_binary[1]}")
# 输出示例：TN: 3, FP: 1, FN: 1, TP: 3
```
