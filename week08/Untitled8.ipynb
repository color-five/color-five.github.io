```python
def multiclass_confusion_matrix(y_true, y_pred, labels=None):
    """
    计算多分类混淆矩阵。
    参数:
        y_true: list 或 array-like, 真实标签。
        y_pred: list 或 array-like, 预测标签。
        labels: list, 可选。指定类别的顺序和名称。如果为None，则使用y_true和y_pred中出现的唯一值并排序。
    返回:
        cm: list of lists, n x n混淆矩阵，其中n为类别数。cm[i][j]表示真实为i类、预测为j类的样本数。
        label_list: list, 实际使用的类别标签列表，与矩阵的行/列顺序对应。
    """
    # 确定类别标签和数量
    if labels is None:
        # 获取所有唯一标签并排序，以确保顺序稳定
        label_list = sorted(list(set(y_true) | set(y_pred)))
    else:
        label_list = labels
    
    num_classes = len(label_list)
    
    # 创建标签到索引的映射字典，便于快速查找
    label_to_index = {label: idx for idx, label in enumerate(label_list)}
    
    # 初始化n x n零矩阵
    cm = [[0 for _ in range(num_classes)] for _ in range(num_classes)]
    
    # 遍历并填充矩阵
    for true_label, pred_label in zip(y_true, y_pred):
        i = label_to_index[true_label]
        j = label_to_index[pred_label]
        cm[i][j] += 1
    
    return cm, label_list

# 示例使用（三分类）
y_true_multi = [0, 1, 2, 2, 2, 0, 1, 1, 0, 1]
y_pred_multi = [0, 2, 2, 2, 0, 0, 1, 1, 0, 2]
cm_multi, used_labels = multiclass_confusion_matrix(y_true_multi, y_pred_multi)
print(f"多分类混淆矩阵 (类别顺序: {used_labels}):")
for row in cm_multi:
    print(row)
# 输出示例（矩阵）:
# [2, 0, 1]  # 真实类别0：2个预测为0，0个预测为1，1个预测为2
# [0, 2, 1]  # 真实类别1：0个预测为0，2个预测为1，1个预测为2
# [1, 0, 2]  # 真实类别2：1个预测为0，0个预测为1，2个预测为2
```
