```python
import numpy as np

def matrix_multiply_numpy(A, B):
    """
    使用NumPy实现矩阵乘法。
    Args:
        A: 类数组结构，可转换为np.ndarray，形状 (m, n)
        B: 类数组结构，可转换为np.ndarray，形状 (n, p)
    Returns:
        C: np.ndarray，乘积矩阵，形状 (m, p)
    """
    A_np = np.array(A)
    B_np = np.array(B)
    
    # 方法1: 使用 np.dot
    C_dot = np.dot(A_np, B_np)
    # 方法2: 使用 @ 运算符 (Python 3.5+)
    C_at = A_np @ B_np
    # 方法3: 使用 np.matmul
    C_matmul = np.matmul(A_np, B_np)
    
    # 验证三种方法结果一致
    assert np.allclose(C_dot, C_at) and np.allclose(C_at, C_matmul), "NumPy计算结果不一致"
    return C_dot

# 测试
A_np = np.array([[1, 2], [3, 4]])
B_np = np.array([[5, 6], [7, 8]])
C_np = matrix_multiply_numpy(A_np, B_np)
print("NumPy计算结果：\n", C_np)  # 输出：[[19 22], [43 50]]
```
