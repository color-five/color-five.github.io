```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import OrdinalEncoder

# ============ 1. 准备数据 ============
data = [
    ("青绿", "蜷缩", "好瓜"),
    ("乌黑", "蜷缩", "好瓜"),
    ("青绿", "硬挺", "坏瓜"),
    ("浅白", "蜷缩", "坏瓜"),
    ("乌黑", "稍蜷", "好瓜"),
    ("青绿", "稍蜷", "好瓜"),
    ("浅白", "硬挺", "坏瓜"),
    ("乌黑", "硬挺", "坏瓜"),
]

X_raw = [row[:2] for row in data]
y = [row[2] for row in data]

# ============ 2. 特征编码（文字→数字）============
enc = OrdinalEncoder()
X = enc.fit_transform(X_raw)

# ============ 3. 训练 ============
model = MultinomialNB()
model.fit(X, y)

# ============ 4. 预测 ============
test_x = enc.transform([["青绿", "蜷缩"]])
pred = model.predict(test_x)
proba = model.predict_proba(test_x)

print(f"预测结果: {pred[0]}")
print(f"各类别概率: {dict(zip(model.classes_, proba[0]))}")
```
