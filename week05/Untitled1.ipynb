```python
import numpy as np 
import matplotlib. pyplot as plt 

np.random.seed(19680801)
#设定随机数种子，保证程序运行结果的可重复性

N=50
x = np. random. rand(N) # 50个点的x坐标
y= np. random. rand(N) #50个点的y坐标
colors = np.random. rand(N) # 50个点的颜色
area=(30 * np.random. rand(N))**2#50个点的面积

plt. scatter(x, y, s = area, c= colors, alpha = 0.5) #显示散点图
plt.show() 
```

```python
import numpy as np 
import matplotlib. pyplot as plt 
import matplotlib 

font = {'family': 'Vera Sans', 'weight': 'bold', 'size':14}# 字体属性
matplotlib. rc('font',**font) # 使用 font 字体

x1=[1,2,3,4,5] 
y1 =[1,4,9,16,25] 
x2=[1,2,3,4,5] 
y2=[2,5,11,12,17] 
#图形为折线图，颜色为red(r)，线型为--，即虚线，星型marker(*) 
plot1 = plt. plot(x1, y1, 'r--*', label = 'red line') 
# 图形为散点图(0)，颜色为green(g)，
plot2 = plt. plot(x2, y2, 'go', label = 'green circles') 
plt.title('Plot of y vs. x',** font)#图的标题，使用font 字体
plt.xticks((1,2,3,4,5),('one','two','three','four','five')) #设置x轴的刻度
plt.xlabel('x axis',**font)#设置x轴标题，使用font字体
plt.ylabel('y axis',** font)#设置y轴标题，使用font字体plt.xlim(0.0,9.0) #设置x轴坐标上下界plt.ylim(0.0,30.)#设置y轴坐标上下界
plt. legend(loc = 'best', shadow = True, ncol = 2) #显示图例plt.show()
```
